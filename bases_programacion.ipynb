# Tipos de datos básicos:
- Basicos
  - Numericos: (Python: Entero, Flotante)
    - ``entero = 5``
    - ``flotante = 5.0`` (que es un flotante)
  - Textuales:
    - ``texto = 'hola neuroco'``
    - caracter : ``'c'``
    - String: cadena de caracteres : ``"njsakdks"``
  - Lógicos: Booleanos
    - ``hoyEsSabado = True``
  - Colecciones: Arreglos (Python: Listas)
- Agrupamiento de datos
  - Listas
    - Ejemplo: lista de nombres de los asistentes
    - ``lista = ['Yorguin','Daniel','Alberto']``
    - Posicionamiento e indexación
      - Zero-based numbering
      - ``print(lista[0])``
      - ``print(lista[-1])``
      - ``print(lista[-3])``
      - ``print(lista[-4])`` --> Error!
  - Conjuntos
    - El conjunto guarda elementos unicos y no ordenados.
    - ['a','a','b','b] --> {'a','b'}
    - ``print(set(['a','a','b','b]))``
  - Mapeos: Diccionario
    - Como agrupar datos como si fueran un formulario
      - crear un diccionario de su identidad
      - ``tu = {'nombre':'Alberto','edad':25,'email':'no recuerdo'}``
      - ``print(tu)``
      - ``print(tu['nombre'])``
